# Financial Ratio Analysis

This notebook analyzes financial data by calculating profitability and leverage ratios and visualizing the relationship between them for specific company types.


```python
# Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data

income_statement = pd.read_excel("data/Income_Statement.xlsx")
balance_sheet = pd.read_excel("data/Balance_Sheet.xlsx")
```

```python
# Merge data
df_ratios = pd.merge(income_statement, balance_sheet, on=["Year", "company", "comp_type"])
```

```python
# Compute ratios
# Compute Gross Margin Ratio
df_ratios["profitability_ratio"] = (df_ratios["Total Revenue"] - df_ratios["Cost Of Goods Sold"]) / df_ratios["Total Revenue"]

# Optional: Compute Operating Margin Ratio (commented out)
# df_ratios["profitability_ratio"] = (df_ratios["Total Revenue"] - df_ratios["Total Operating Expenses"]) / df_ratios["Total Revenue"]

# Compute Debt-to-Equity Ratio
df_ratios["leverage_ratio"] = df_ratios["Total Liab"] / df_ratios["Total Stockholder Equity"]
```

```python
# Analyze by company type
# Compute Gross Margin Ratio
df_ratios["profitability_ratio"] = (df_ratios["Total Revenue"] - df_ratios["Cost Of Goods Sold"]) / df_ratios["Total Revenue"]

# Compute Debt-to-Equity Ratio
df_ratios["leverage_ratio"] = df_ratios["Total Liab"] / df_ratios["Total Stockholder Equity"]

```

```python
# Create visuals
# Filter for real estate companies
df_real_est = df_ratios[df_ratios["comp_type"] == "real_est"]

# Plot leverage vs. profitability
sns.regplot(data=df_real_est, x="leverage_ratio", y="profitability_ratio")
plt.title("Leverage vs. Profitability for Real Estate Companies")
plt.xlabel("Leverage Ratio (Debt-to-Equity)")
plt.ylabel("Profitability Ratio (Gross Margin)")
plt.tight_layout()
plt.show()
```
